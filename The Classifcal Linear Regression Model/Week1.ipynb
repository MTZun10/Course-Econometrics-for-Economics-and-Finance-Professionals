# Types of Data
- cross sectional data:  the situation where the time is fixed and the observation changes. (company size and firm's perfomance, unemployment and growth) - linear relationship
- time series data: For time series data you fix the individual and you change the point in time at which you study the individual. (GDP over time)
- panel data (longitudinal data): a combination of crossectional and time series data - balanced & unbalanced panel data.
- big data: has less or no structure

** Detrending: Differencing: Subtracting the current value from the previous value (y(t)-y(t-1)) to remove linear trends.

** High frequency data: for which the intervals when the data is gathered are very small. (financial data)

# Example


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
month = list(np.arange(1,5))
excess_return = [17.8, 39.0, 12.8, 24.2, 17.2]
excess_return_market_index = [13.7, 23.2, 6.9, 16.8, 12.3]

plt.scatter(excess_return_market_index, excess_return)
plt.xlabel("Excess Retrun on Market Portfolio")
plt.ylabel("Excess Return on Fund")
plt.show();
```

```python
y = np.array(excess_return).reshape(-1, 1)
X = np.array(excess_return_market_index).reshape(-1, 1)


model = LinearRegression()
model.fit(X, y)

print(f"Intercept (alpha): {model.intercept_[0]:.2f}")
print(f"Slope (beta): {model.coef_[0][0]:.2f}")
```
